```python
import numpy as np
alt =np.random.normal(1.75, 0.30, 2000)
alt
```

```python
alt = alt.tolist()
```

```python
from sympy import *
from random import uniform
x1 = symbols('alt_i')
x2 = symbols('aleatorio')
peso_i = x1*60 + x2
peso_i
```

```python
peso = []
for i in alt:
  a = round(float(uniform(-5, 5)), 2)
  b = 50*i
  c = a + b
  peso.append(c)
```

```python
len(peso)
```

```python
peso1, altura1 = symbols(['peso', 'altura'])
IMC_formula = (peso1 / altura1**2)
IMC_formula
```

```python
IMC = []
for i, j in zip(peso,alt):
  e = i / (j**2)
  e = round(e, 2)
  IMC.append(e)
```

```python
len(IMC)
```

```python
SEXO = []
for i in range(0,2000):
  value = np.random.uniform(0,1)
  if value > 0.5:
    SEXO.append('M')
  else:
    SEXO.append('F')
```

```python
SEXO
```

```python
CARDIACO = []
for i in range (0,2000):
  if IMC[i] > 40:
    CARDIACO.append('S')
  else:
    if IMC[i] > 30 and IMC[i] < 40:
      CARDIACO.append('P')
    else:
        CARDIACO.append('N')
```

```python
z = alt,peso, IMC, SEXO, CARDIACO
```

```python
import pandas as pd
df = pd.DataFrame(z, index=['Altura', 'Peso', 'IMC', 'SEXO', 'CARDIACO']).T
```

```python
df
```

```python
def media(tabela):
  soma = sum(tabela)
  media = soma / len(tabela)
  media =round(media, 2)
  print(media)
```
